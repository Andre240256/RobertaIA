```python
import gymnasium as gym
from gymnasium.envs.registration import register
from robertaEnv import RobertaEnv

register(
    id='RobertaEnv-v0',
    entry_point='robertaEnv:RobertaEnv',
    max_episode_steps=10000,
)
```

```python
from gymnasium.utils.env_checker import check_env

env = gym.make('RobertaEnv-v0')
check_env(env.unwrapped)
```

```python
from robertaEnv import RobertaEnv
import numpy as np

env = RobertaEnv()
obs, _ = env.reset()
print(f"Passon start: phi={obs[0]:.4f}, vel={obs[1]:.4f}")

for i in range(100):
    obs, reward, terminated, truncated, info = env.step(np.array([0.2], dtype=np.float32))
    print(f"Passon {i}: phi={obs[0]}, vel={obs[1]}")
    


    if obs[0] < -52 or obs[0] > 45:
        print("Obs is out of bounds")
        break
```
